## Generators 

It is a sub class of the iterators . It uses yield keyword to produce a series of values lazily. which means they generate values and fly do not store in memory 


generators are particularly useful for reading large files because they allow you to process one line at a time without loading the entire file into memory

```python
def sqaure(n) :
    for i in range(n) :
        yield i*i
```

```python
sqaure(3)
```

```python
for i in sqaure(4) :
    print(i)
```

```python
a=sqaure(3)
```

```python
print(a)
```

```python
next(a)
```

```python
def my_generator() :
    yield 1
    yield 2 
    yield 3
```

```python
gen=my_generator()
print(gen)
```

```python
next(gen)
```

```python
for val in gen :
    print(val)
```

```python
### practical example : reading large files 

def read_file(file_path) : 
    with open(file_path,'r') as file :
        for line in file :
            yield line 
    
```

```python
textFile=read_file("dhoni.txt") 
next(textFile)
```

```python
for line in textFile :
    print(line)
```
