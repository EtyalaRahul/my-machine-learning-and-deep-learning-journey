### SQL AND SQLITE

SQL(Structured query language) is a standard language for managing and manipulating relational databases. SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems.

```python
import sqlite3 
```

```python
## connect to an sqlite database 
connection=sqlite3.connect("example.db")
print(connection)
```

```python
cursor=connection.cursor()
```

```python
## create a table
cursor.execute("""CREATE TABLE IF NOT EXISTS users
                  (id INTEGER PRIMARY KEY,
                  name TEXT, 
                  age INTEGER,
                  department text)
                  """)
connection.commit()
```

```python
cursor.execute("""select * from users;
                  """)
```

```python
cursor.execute("""
INSERT INTO users (name, age, department) VALUES
('Rahul', 22, 'CSE'),
('Pavan', 23, 'ECE'),
('Teja', 21, 'MECH'),
('Siri', 22, 'IT'),
('Vinay', 24, 'EEE'),
('Hemanth', 25, 'CIVIL'),
('Kiran', 23, 'AIML'),
('Priya', 22, 'CSE'),
('Neha', 21, 'IT'),
('Arjun', 24, 'CSE')
""")

connection.commit()
```

```python
cursor.execute('select * from users;')
print(cursor.fetchall())
```

```python
import pandas as pd

df = pd.read_sql("SELECT * FROM users", connection)
df
```

```python
## querying the data from sqlite3 
cursor.execute('''
               select * from users;''')
rows=cursor.fetchall()
for row in rows :
    print(row)
```

```python
cursor.execute("""
UPDATE users
SET age = 23, department = 'IT'
WHERE id = 1
""")
connection.commit()
```
