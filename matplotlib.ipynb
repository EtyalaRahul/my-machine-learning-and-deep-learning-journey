### Data visualization with matplotlib


Matplotlibe is a powerful plotting library for python that enables the creation of static, animated, and interactive visualizations. It is widely used for data science and analysis.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create a line plot 
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis') 
plt.title('Basic line plot ')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25] 


## customized line plot  
plt.plot(x,y,color='red',linestyle='--',marker='*')
```
