```python
class Node :
    def __init__(self,val) :
         self.data=val
         self.next=None
```

```python
## conversion of array to linked list 

def arr_to_linkedlist(arr) :
    if len(arr)==0 :
        return 
    head=Node(arr[0])
    curr=head
    for i in range(1,len(arr)) :
        temp=Node(arr[i])
        curr.next=temp
        curr=curr.next 
    return head 
```

```python
##print linked list 
def printLL(head) :
    if head is not None :
        temp=head 
        while temp is not None :
            print(temp.data)
            temp=temp.next 
    
```

```python
## length of the linkedlist 
def lengthofLL(head) :
    if head is not None :
        cnt=0 
        temp=head 
        while temp is not None :
            cnt+=1 
            temp=temp.next 
        return cnt 
    else :
        return 0
```

```python

n=int(input())
lst=list(map(int,input().split()))
head=arr_to_linkedlist(lst)
printLL(head) 
```

```python
lengthofLL(head)
```

```python
## search in a linked list 

def search(head,val) :
    if head is not None :
        temp=head 
        flag=False
        while temp is not None :
            if temp.data ==val :
                flag=True 
                break 
            temp=temp.next
        return flag
    else :
        return False
        
```
