2. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
import numpy as np

# Define a structured data type with two integer fields: x and y
data_type = [('x', 'i4'), ('y', 'i4')]

# Create a structured NumPy array using the above data type
data = np.array([(1, 2), (3, 4), (5, 6)], dtype=data_type)

print("Original Array:\n")
print(data)

print("\nEuclidean distance:\n")

# Compute pairwise Euclidean distance between all points
# Explanation:
# data['x'] gives all x-coordinates → [1, 3, 5]
# data['y'] gives all y-coordinates → [2, 4, 6]
#
# data['x'][:, np.newaxis] converts shape from (3,) to (3,1)
# so broadcasting can happen with data['x'] which is (3,)
#
# Formula: distance = sqrt( (x1 - x2)^2 + (y1 - y2)^2 )
# Broadcasting automatically computes this for all pairs
distance = np.sqrt(
    (data['x'][:, np.newaxis] - data['x']) ** 2 +
    (data['y'][:, np.newaxis] - data['y']) ** 2
) 

print(distance)
```
