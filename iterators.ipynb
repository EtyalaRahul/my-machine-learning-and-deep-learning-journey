# iterators

#### iterator uses lazy loading technique 

##### next() to access the elements in the iterator

why iterators ? 

Iterators are advanced python concepts that allows efficient looping and memory management. Iterators provide and way to access elements of a collection sequentially without exposing the underlying structure


```python
mylist=[1,2,3,4,5,6]
for i in mylist :
    print(i)
```

```python
type(mylist)
```

```python
print(mylist)
```

```python
###iterator 
iterator=iter(mylist)
```

```python
type(iterator)
```

```python
next(iterator)
```

```python
##stop iteration exception is not stop iteration in the try catch block 



try :
    print(next(iterator)) 
    
except  StopIteration :
    print("there are no more elements present in the list")
```
