### Skill 1

Implement an image-quality enhancement pipeline that applies adaptive histogram 
equalization followed by bilateral filtering and dynamic Otsu-thresholding. Compare the 
preprocessing results on 5 different noisy images

```python

#importing necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_paths = [
    "img1.jpg",
    "img2.jpg",
    "img3.jpg",
    "img4.jpg",
    "img5.jpg"
]

# Create CLAHE object
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Define the processing pipeline

''' 
The pipeline takes a grayscale image as input.  
CLAHE is applied to improve local contrast in different regions of the image.  
Contrast limiting prevents excessive noise amplification.  
The enhanced image may still contain noise after contrast adjustment.  
A bilateral filter is used to smooth the image.  
This filter reduces noise while preserving important edges.  
Edge preservation is crucial for accurate segmentation.  
Otsus thresholding is then applied to the filtered image.  
It automatically selects an optimal threshold value.  
The final output is a clean binary segmented image.


'''
def pipeline(image):
    clahe_img = clahe.apply(image)
    bilateral_img = cv2.bilateralFilter(clahe_img, 9, 75, 75)
    _, otsu_img = cv2.threshold(
        bilateral_img, 0, 255,
        cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )
    return clahe_img, bilateral_img, otsu_img

images = [cv2.imread(p, cv2.IMREAD_GRAYSCALE) for p in image_paths]

# Process and visualize each image
for i, img in enumerate(images):
    clahe_img, bilateral_img, otsu_img = pipeline(img)

    plt.figure(figsize=(12, 3))

    plt.subplot(1, 4, 1)
    plt.imshow(img, cmap="gray")
    plt.title("Original")
    plt.axis("off")

    plt.subplot(1, 4, 2)
    plt.imshow(clahe_img, cmap="gray")
    plt.title("CLAHE")
    plt.axis("off")

    plt.subplot(1, 4, 3)
    plt.imshow(bilateral_img, cmap="gray")
    plt.title("Bilateral")
    plt.axis("off")

    plt.subplot(1, 4, 4)
    plt.imshow(otsu_img, cmap="gray")
    plt.title("Otsu")
    plt.axis("off")

    plt.suptitle(f"Image {i+1}")
    plt.show()
```
