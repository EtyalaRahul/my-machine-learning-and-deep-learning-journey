### Data visualization with seaborn 

Seaborn is a python visualization library based on mathplotlib that provides a high-level interface for drawing attractive and informative stastical graphics. Seaborn helps in creating complex visualizations with just a few linse of code

```python
import seaborn as sns
```

```python
## basic plotting with seaborn 
tips=sns.load_dataset('tips') 
tips
```

```python
tips.describe()
```

```python
## create a scatter plot using tips data 
import matplotlib.pyplot as plt 
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scatter plot of total bill vs tip")
```

```python
## line plot 

sns.lineplot(x='size',y='total_bill',data=tips) 
plt.title("Line plot of Total bill by size")
```

```python
## categorial plots 
## bar plots

sns.barplot(x='day',y='total_bill',data=tips,color='pink')
plt.title("Bar plot of total bill by day")
```

```python
## Box plot 

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
##violin plot 
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## histograms 
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False)
```

```python
## kde plot 

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## pair plot 
sns.pairplot(tips)
```

```python
## heat map
corr=tips[['total_bill','tip','size']].corr() 
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
