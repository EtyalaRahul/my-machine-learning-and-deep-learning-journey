```python
import requests

url = "https://en.wikipedia.org/wiki/Mobile_country_code"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)
response.raise_for_status()   # will show errors if any

tables = pd.read_html(response.text, match="Country")
len(tables), tables[0].head()
```

## what is pickle file 

A pickle file is a file created using Pythons pickle module, which is used to serialize and deserialize Python objects.

 What does that mean?
 Serialization (Pickling)

Converts a Python object (like a list, dictionary, model, class object, etc.) into a binary format so it can be saved to a file or sent over a network.

 Deserialization (Unpickling)

Converts the binary data back into the original Python object.

```python
df_excel.to_pickle("rahul")
```

```python
## conversion of pickle file to normal or reading data from the pickle file 

pd.read_pickle("rahul")
```
