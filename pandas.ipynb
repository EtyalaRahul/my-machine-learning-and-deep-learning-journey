### pandas

pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structures. Series and DataFrame. A series is a one dimensional array like object, while data frame is a two dimensional, size-mutable, and potentially heterogenous tabulat data structures while labelled axes(rows and columns) 


what is pandas series ? 

It is a one dimensional array like objects  that can hold any data type. It is similar to a column in a table
